# Size of a DataFrame


# The **<font color ='brown'>size, shape, and len</font>** methods

We should always check the data’s size before analyzing it. The size of a <font color ='red'>DataFrame</font> can be expressed in terms of the number of rows and columns.


![Screenshot 2024-04-17 at 8.40.41 AM.png](<attachment:Screenshot 2024-04-17 at 8.40.41 AM.png>)

The <font color ='red'>shape</font> method returns a <u>tuple that contains the number of rows and columns</u>. The <font color ='red'>size</font> method contains a number that shows <u>the number of rows multiplied by the number of columns</u>. Thus, it returns the total number of cells in a <font color ='red'>DataFrame</font>. The built-in Python function, <font color ='red'>len</font> gives us <u>the number of rows</u> in a <font color ='red'>DataFrame</font>. Let’s check the size of the sales using these methods.

```python
import pandas as pd

sales = pd.read_csv("sales.csv")

print(sales.shape)
```

```python

print(sales.size)
```

```python
print(len(sales))
```

If we’re only interested in the number of rows in a <font color ='red'>DataFrame</font>, then the <font color ='red'>len</font> function works. The <font color ='red'>shape</font> method returns both the number of rows and columns in a tuple. We rarely use the <font color ='red'>size</font> method.
